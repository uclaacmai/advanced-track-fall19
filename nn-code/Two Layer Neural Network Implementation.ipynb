### Vectorization: A primer
Vectorization is an important tool to understand when writing machine learning code, especially when you're implementing your own loss functions, gradient descent algorithm, and not relying on high-level librariers. Proper vectorization can reduce the runtime of code from days to hours or from hours to minutes.

#### Why do we need vectorization at all?

- We can define "vectorization" concretely as taking an operation that works on a set of values one-by-one, to working on them all at once on the same time. 
- If you consider the CPU-level of your code being instructed, there are roughly two classes of instruction: Single-Instruction Single data, and Single-Instruction Multiple Data. With vectorization, we can apply a single instruction to multiple data, which is clearly more efficient than applying it over and over again to single data.
- In machine learning, we are often dealing with function application to matrix data, for example $ M * N $ data where we have $M$ instances of feature vectors of size $ N $. Applying a vectorized version of, for example, a sigmoid function, to every single element would be a lot less efficient than applying it to the whole data.
